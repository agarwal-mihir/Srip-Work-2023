```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = xr.open_dataset('data/era5_data2.grib', engine = 'cfgrib')
data
```

```python
df = data.to_dataframe().reset_index()
df
```

```python
uniq = df.drop_duplicates(subset = ['latitude', 'longitude'])[['latitude', 'longitude']]
```

```python
uniq
```

```python
df['time'] = pd.to_datetime(df['time'])

from datetime import timedelta
def shift_time(x):
    x['time'] = x['time'] + timedelta(minutes = 30)
    return x

df = df.apply(lambda x: shift_time(x), axis = 1)

df
```

```python
df[['time', 'longitude', 'latitude', 'u10', 'v10', 't2m', 'msl']].to_csv('wind_temp2.csv')
```
